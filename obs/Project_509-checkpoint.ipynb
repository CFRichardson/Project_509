```python
from yelpapi import YelpAPI
yelp_api = YelpAPI(api_key)
```

## Search criteria

```python
term = 'restaurant'
location = 'San Diego, CA'
search_limit = 50

response = yelp_api.search_query(term = term,
                                 location = location,
                                 limit = search_limit)

type(response)
```

```python
cols = list(response['businesses'][0].keys())
data = pd.DataFrame(columns=cols)
            
for biz in response['businesses']:
    data = data.append(biz, ignore_index=True)
            
data.head()
```

## Web scraping reviews

```python
restaurant = []

for row in range(0,len(data)):
    # got this from https://proxiesapi.com/blog/scraping-yelp-data-with-python-and-beautiful-soup.html.php
    url=pd.DataFrame(data)['url'][row]
    response=requests.get(url)
    soup=BeautifulSoup(response.content,'lxml')

    # got this from https://discuss.dizzycoding.com/how-to-get-text-from-span-tag-in-beautifulsoup/
    spans = soup.findAll('span', attrs = {'class' : 'raw__09f24__T4Ezm'})

    for span in spans:
        restaurant.append({'restaurant': pd.DataFrame(data)['name'][row]
                         , 'reviews': span.text})

reviews = pd.DataFrame(restaurant)
reviews.sample(5)    
```
